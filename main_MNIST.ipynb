# MNIST Database

```python
#Packages

from sklearn.datasets import load_digits
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load class of the MNIST dataset
mnist = load_digits()

# Class type of the DB
print(type(mnist))

# Attributes of the class
print(mnist.keys())

# Data of the images (Arrays): 64 dimensions
pd.DataFrame(mnist.data).head()
data=pd.DataFrame(mnist.data)
print(data.head())
```

```python
print(data.shape)

# Ground truth of the data
Ground_truth=pd.DataFrame(mnist.target)
print(Ground_truth)
```

```python
print(mnist.images)
plt.imshow(mnist.images[0])
```

```python
fig, axes = plt.subplots(3, 10, figsize=(16, 6))
for i in range(30):
    axes[i//10, i %10].imshow(mnist.images[i], cmap='gray')
    axes[i//10, i %10].axis('off')
    axes[i//10, i %10].set_title(f"target: {mnist.target[i]}")
    
plt.tight_layout()
```

```python
print(mnist.frame)

# Image 8x8
print(mnist.feature_names)

# Names
print(mnist.target_names)

# Description
print(mnist.DESCR)
```

#### Standardize the data: PCA requires that the data is standardized, meaning that each feature has zero mean and unit variance. We can achieve this by using the StandardScaler from the scikit-learn library.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_std = scaler.fit_transform(data)
```

#### Apply PCA: We can use the PCA function from the scikit-learn library to apply PCA to the standardized data. The n_components parameter specifies the number of principal components to keep. Here, we will keep 50 principal components.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
print(pca)
data_pca = pca.fit_transform(data_std)
print(data_pca)
```

#### Visualize the results: We can visualize the results by plotting the first two principal components against each other, colored by their corresponding class labels. This will give us an idea of how well the data can be separated using the reduced number of dimensions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar índices aleatorios
random_indices = np.random.choice(len(data_pca), size=150, replace=False)

# Obtener los datos correspondientes a los índices aleatorios
data_pca=np.array(data_pca.tolist())
new_ground_truth=np.array(mnist.target)
Ground_truth=new_ground_truth
data_pca_random = data_pca[random_indices]
Ground_truth_random = Ground_truth[random_indices]

cmap = plt.get_cmap('jet', 20)
cmap.set_under('gray')

# Graficar los datos
plt.scatter(data_pca_random[:, 0], data_pca_random[:, 1], c=Ground_truth_random, cmap=cmap)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar()
plt.show()

```

```python
# 3 first principal components

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(data_pca_random[:, 0], data_pca_random[:, 1], data_pca_random[:, 2], c=Ground_truth_random, cmap='viridis')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

# Limit plot
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_zlim(-4, 4)

plt.show()
```

#### After applying PCA to the MNIST dataset, we can use the reduced feature set to train a classifier. Here is an example of how to train a logistic regression classifier on the reduced MNIST dataset

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_pca, Ground_truth, test_size=0.2, random_state=42)

# Train a logistic regression classifier on the training data
clf = LogisticRegression()
clf.fit(X_train, y_train)

# Evaluate the classifier on the testing data
accuracy = clf.score(X_test, y_test)
print('Accuracy: {:.2f}%'.format(accuracy*100))

# from 64 until 49 to not lose accuracy rate.
```

#### To visualize the effect of the number of principal components on the accuracy of the logistic regression model, you can create a line plot where the x-axis represents the number of principal components and the y-axis represents the accuracy of the logistic regression model.

```python
import warnings
warnings.filterwarnings('ignore')
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data, Ground_truth, test_size=0.2,random_state=42)

# Define the range of principal components to test
n_components = np.arange(2, 64)

# Fit logistic regression models with varying numbers of principal components
accuracy_scores = []
for n in n_components:
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    logreg = LogisticRegression()
    logreg.fit(X_train_pca, y_train)
    accuracy_scores.append(logreg.score(X_test_pca, y_test))

# Create a line plot of accuracy scores vs. number of principal components
plt.figure(figsize=(17, 10))
plt.plot(n_components, accuracy_scores)
plt.xlabel('Number of principal components')
plt.ylabel('Accuracy score')
plt.xticks(np.arange(2, 64, 2))  # Set x-tick values to show increments of 5
plt.title('Accuracy of Logistic Regression with PCA')
plt.grid()  # Add a grid to the plot
plt.show()
```

#### PCA Analysis: 2 components

```python
data=pd.DataFrame(mnist.data)
Ground_truth=pd.DataFrame(mnist.target)
from sklearn.preprocessing import MinMaxScaler

scaler = StandardScaler()
data_std = scaler.fit_transform(data)

#data_std=data

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_std)

# Generar índices aleatorios
random_indices = np.random.choice(len(data_pca), size=len(data_pca), replace=False)

# Obtener los datos correspondientes a los índices aleatorios
data_pca=np.array(data_pca.tolist())
new_ground_truth=np.array(mnist.target)
Ground_truth=new_ground_truth
data_pca_random = data_pca[random_indices]
Ground_truth_random_1 = Ground_truth[random_indices]

cmap = plt.get_cmap('jet', 30)
cmap.set_under('gray')

# Graficar los datos
plt.scatter(data_pca_random[:, 0], data_pca_random[:, 1], c=Ground_truth_random_1, cmap=cmap)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar()
plt.show()
```

#### MDS: 2 components

```python
from sklearn.manifold import MDS

data=pd.DataFrame(mnist.data)
Ground_truth=pd.DataFrame(mnist.target)

scaler = MinMaxScaler()
data_std = scaler.fit_transform(data)

mds = MDS(n_components=2)
data_mds = mds.fit_transform(data_std)

# Generar índices aleatorios
random_indices = np.random.choice(len(data_mds), size=len(data_mds), replace=False)

# Obtener los datos correspondientes a los índices aleatorios
data_mds=np.array(data_mds.tolist())
new_ground_truth=np.array(mnist.target)
Ground_truth=new_ground_truth
data_mds_random = data_mds[random_indices]
Ground_truth_random_2 = Ground_truth[random_indices]

cmap = plt.get_cmap('jet', 30)
cmap.set_under('gray')

# Graficar los datos
plt.scatter(data_mds_random[:, 0], data_mds_random[:, 1], c=Ground_truth_random_2, cmap=cmap)
plt.xlabel('MDS1')
plt.ylabel('MDS2')
plt.colorbar()
plt.show()
```

#### T-SNE: 2 components

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_std = scaler.fit_transform(data)
#data_std = data/max(data)

tsne = TSNE(n_components=2, random_state=42,perplexity=20)
data_tsne = tsne.fit_transform(data_std)

# Generar índices aleatorios
random_indices = np.random.choice(len(data_tsne), size=len(data_tsne), replace=False)

# Obtener los datos correspondientes a los índices aleatorios
data_tsne_random = data_tsne[random_indices]
Ground_truth_random_3 = Ground_truth[random_indices]

# Graficar los datos
cmap = plt.get_cmap('jet', 30)
cmap.set_under('gray')
plt.scatter(data_tsne_random[:, 0], data_tsne_random[:, 1], c=Ground_truth_random_3, cmap=cmap)
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.colorbar()
plt.show()
```
